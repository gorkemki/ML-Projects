```python

import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
# #############################################################################
# Generate sample data
# Importing the dataset
dataset = pd.read_csv('company_dataxs.csv')

dataset.head()
dataset.columns
X = dataset.iloc[:, :-1].values # Independent Variables
y = dataset.iloc[:, -1].values   # Dependent Variables


# Feature Scaling
from sklearn.preprocessing import  MinMaxScaler
sc= MinMaxScaler()
X= sc.fit_transform(X)
y= y.reshape(-1,1)
y=sc.fit_transform(y)


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=0)  


# Random Forest Regression
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=200, random_state=0)  
regressor.fit(X_train, y_train)  
y_pred = regressor.predict(X_test)  

from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  



# Support Vector Machine - Model 1
regressor=SVR()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)

print('SVR1 Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('SVR1 Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('SVR1 Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  


# Support Vector Machine - Model 2
regressor=SVR(kernel='rbf',epsilon=1.0)
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)

print('SVR2 Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('SVR2 Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('SVR2 Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  



# Support Vector Machine - Model 3
regressor=SVR(kernel='poly',epsilon=1.0)
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
 

print('SVR3 Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('SVR3 Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('SVR3 Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))  

    

y_pred=y_pred.reshape(-1, 1)
y_test=y_test.reshape(-1, 1)
inv_y_pred = sc.inverse_transform(y_pred) # Rescaling back
inv_y_test = sc.inverse_transform(y_test)
print('Mean absolute Error Rev:', metrics.mean_absolute_error(inv_y_test, inv_y_pred)) 


```
